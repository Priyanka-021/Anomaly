**Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

  Answer: Anomaly Detection (also known as outlier detection) is the process of identifying data points that deviate significantly from the majority of the data. These anomalies may indicate: Fraudulent activity, Faulty equipment,Data quality issues, Unusual patterns in system behavior. It is widely used in areas like: Banking (fraud detection), Cybersecurity (intrusion detection),
  Manufacturing (fault detection),Healthcare (disease outbreak or medical errors)
  Types of Anomalies- Anomalies can be classified into three main types:
  1.Point Anomaly (Global Outlier)- A single data point is significantly different from the rest of the dataset.
  Example: In a dataset of daily temperatures: [30°C, 29°C, 31°C, 100°C, 30°C]
  → 100°C is a point anomaly — it's too high to be normal.
  Use Case: Credit card fraud: A transaction of ₹1,00,000 when usual transactions are ₹500-₹1000.
  2.Contextual Anomaly (Conditional Outlier)- A data point is only anomalous in a specific context or condition (e.g., time, location).
  Example: Temperature of 20°C: Normal in March (spring), Anomalous in December (winter) Here, time is the context.
  Use Case: Stock prices: A 10% drop in one day might be normal during a crisis but not during a stable period.
  Heart rate: 100 bpm during exercise is fine, but abnormal while sleeping.
  3.Collective Anomaly (Group Outlier)- A group of data points is anomalous together, even if individual points are not.
  Example: A series of login attempts within a second: Each login attempt may look normal. Together, they may indicate a brute-force attack.
  Use Case: Network traffic: A sudden burst of packets from a device could indicate a DDoS attack.
  Sensor data: A group of pressure readings forming an unusual pattern may indicate a machine failure.

  **Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

  Isolation Forest-:
  Approach:Builds a forest of randomized decision trees that recursively partition the feature space. Outliers—being rare and different—are easier to isolate, so they appear near the top of trees with shorter path-lengths
  Strengths: Linear time complexity and low memory usage → highly scalable to large datasets.
  Works well in high-dimensional feature spaces without explicit distance metrics or density estimation
  Weaknesses: May struggle when anomalies aren’t clearly isolated (“swamping” or “masking” issues)
  Less effective for detecting local anomalies in varying-density datasets
  Best Use Cases:
  Large-scale anomaly detection (e.g. millions of data points)
  High-dimensional data streams (e.g. sensor data, metrics, financial transactions)
  Production monitoring and real-time flagging systems Towards Data Science
  
  Local Outlier Factor (LOF)
  Approach: Estimates local data density via k-nearest neighbors. Compares the density of a point to that of its neighbors. If its local density is significantly lower, it's flagged as an outlier
  Strengths: Excellent at identifying local anomalies in regions with varying density
  Interpretable — provides an LOF score (ratio of neighbor densities). Values 1 typically indicate outliers
  Shares conceptual similarities with DBSCAN (like core distance, reachability) so it's good for structured, spatial data
  Weaknesses: Computationally expensive with O(n²) nearest neighbor searches, especially in high dimensions
  Sensitive to parameter choice (k value) and data scaling; threshold interpretation can vary across datasets
  Best Use Cases: Smaller or moderate-sized datasets with local density variations.
  Applications needing high precision in anomaly detection, such as healthcare diagnostics, fraud investigation, manufacturing QA, or scientific analysis

  DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
  Approach: Clusters data based on density (points with ε-neighbors ≥ MinPts are “core points”). Points not in any dense region are labeled as noise—interpreted as outliers
  Strengths: Doesn't require specifying the number of clusters.
  Can detect arbitrarily shaped and sized clusters; naturally labels outliers as noise
  Visually intuitive clusters and noise separation (in lower-dimensional data)
  Weaknesses: Struggles with high-dimensional data—distance measures lose meaning; performance degrades
  Choosing ε and MinPts can be tricky, especially in varied density datasets; outputs sensitive to parameter tuning
  Not ideal for real-time or very large-scale anomaly detection (slower than Isolation Forest)
  Best Use Cases: Moderate-sized datasets with spatial structure or well-defined clusters of differing shapes.
  When you need both clustering and anomaly identification, such as identifying noise in geospatial, sensor, or behavioral data.
  Exploratory analysis where cluster structure and outlier detection are both relevant.

  **Question 3: What are the key components of a Time Series? Explain each with one example.**
   Answer: the key components of a time series, each with a concrete example:
   1.Trend (T)
   Definition: The long-term increase, decrease, or stability in a time series—often nonlinear or linear.
  Example: A company's annual revenue steadily rising from ₹100k in 2019 to ₹210k in 2023 shows an upward trend.
  2.Seasonality (S)
  Definition: Regular, repeating patterns at fixed intervals (e.g. daily, monthly, yearly), driven by calendar effects: festivals, weather, day-of-week cycles.
  Example: Ice-cream sales peaking every summer (May-August) and dropping in winter is a seasonal pattern that repeats annually.
  3.Cyclic Component (C)
  Definition: Long-term fluctuations that repeat irregularly over multi-year periods—not tied to calendar periods, often driven by economic or business cycles.
  Example: GDP growth exhibiting periods of expansion and recession over several years—these cycles vary in duration and timing.
  4.Irregular (Noise) Component (I)
  Definition: Random, unpredictable, non-systematic fluctuations not explained by trend, seasonality, or cycles. These are the residuals after removing structured patterns.
  Example: A sudden drop in the stock market index in 2020 due to the COVID‑19 pandemic—a shock not part of any trend or repeating pattern

  **Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**
  
  Answer: A stationary time series is one whose statistical properties—such as the mean, variance, and autocovariance—remain constant over time. In weak (covariance) stationarity, the mean is constant and the autocovariance depends only on the time lag, not on the actual time index
  In strict stationarity, the entire joint distribution is invariant under shifts in time.
  In trend-stationary series, a deterministic trend may exist, but removing it yields stationarity; shocks are temporary.
  In contrast, a unit-root or difference-stationary series has stochastic trends: shocks have a permanent effect unless differenced
  Test for Stationarity
  1.Visual and statistical checks- Plot the series to look for obvious trends, changing variance, or seasonality.
  Split the data (e.g. first half vs. second half) and compare their means and variances. Significant differences suggest non-stationarity
  2.Unit-root tests
  Augmented Dickey-Fuller (ADF) test:
  Null hypothesis: series has a unit root (non-stationary)
  Rejecting the null implies stationarity
  Phillips-Perron (PP) test: Robust to autocorrelation and heteroskedasticity; similar null hypothesis.
  KPSS test:
  Null hypothesis: series is trend-stationary (stationary around a trend)
  Failing to reject suggests trend-stationarity; rejecting suggests unit-root non-stationarity
  A combination of ADF (or PP) plus KPSS helps triangulate whether data are truly stationary, trend-stationary, or unit-root non-stationary.
   Transform a Non-Stationary Series
  i.Differencing- First-order differencing: removes a unit root trend.
  If still non-stationary, you can apply second differencing.
  Seasonal differencing- subtracting values separated by a seasonal period
  handles seasonal patterns.
  ii.Transformations to stabilise variance- Log, square-root, power, or Box-Cox transformations can stabilize variance in series with increasing variability over time.
  If variance grows with level (e.g. exponentially increasing series), logs are particularly useful
  iii.Detrending by modeling- Fit a linear or nonlinear trend model (e.g. linear regression of y at t vs. time), subtract the fitted trend, and use residuals for analysis
  Other filters like Hodrick-Prescott, Baxter-King, or Christiano-Fitzgerald can isolate trend/cycle components
  iv.Advanced methods- Surrogate data methods (e.g. Fourier-based filtering) generate a version preserving autocorrelation structure but stationarized statistically

  **Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.**

  Answer:a clarified breakdown of AR, MA, ARIMA, SARIMA, and SARIMAX models—highlighting their structure and when to use each:
  1.AR (Autoregressive)
  Structure: Depends on its own past values.
  Use cases: When current values relate linearly to a fixed number of past lags.
  Example: Modeling daily temperature where today's temp depends on the previous 3 days.
  2.MA (Moving Average)
  Structure: Depends on past error terms.
  Use cases: To smooth short-term shocks with a rolling average of past errors.
  Example: Financial returns where today's return is a function of recent shock residuals.
  3.ARIMA (Autoregressive Integrated Moving Average)
  Structure: An ARMA(p,q) model applied after differencing d times to remove trends (integration component).
  ARIMA(p,d, q) = “difference d → model AR(p)+MA(q)”
  Use cases: Non-stationary data with trends but no seasonality.
  Example: Forecasting inflation rates over time where trends exist but seasonality doesn't.
  4.SARIMA (Seasonal ARIMA)
  Structure: Extends ARIMA by adding seasonal terms (P,D,Q) with period m.
  SARIMA(p,d,q)(P,D,Q)_m: handles both trend and repeating seasonal patterns
  Use cases: Series with both trend and clear seasonal cycles.
  Example: Monthly electricity demand forecasting with yearly seasonality (m=12).
  5.SARIMAX (Seasonal ARIMA with eXogenous variables)
  Structure: All of SARIMA's components plus exogenous predictors X_t (e.g., holidays, marketing spend).
  Modeled as SARIMA structure, with added regressors
  Use cases: When external variables influence the target variable alongside trend and seasonality.
  Example: Retail sales forecasting using prices and promotions as regressors, with seasonality and trend.

**Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components (Include your Python code and output in the code box below.)**
Answer:




```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose

# 1. Load the dataset directly from statsmodels
data = sm.datasets.get_rdataset("AirPassengers", "datasets", cache=True).data
data = pd.DataFrame(data["value"]).set_index(
    pd.date_range(start="1949-01", periods=len(data), freq="M")
)
data.columns = ["Passengers"]

# 2. Plot the original time series
plt.figure(figsize=(10, 4))
plt.plot(data, label="Passengers")
plt.title("AirPassengers Time Series")
plt.ylabel("Number of Passengers")
plt.xlabel("Date")
plt.legend()
plt.show()

# 3. Perform classical seasonal decomposition (additive model)
decomp = seasonal_decompose(data["Passengers"], model="additive", period=12)

# 4. Plot decomposition results
fig = decomp.plot()
fig.set_size_inches(12, 8)
plt.show()

# Access individual components if needed
trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid
```

 **Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot. (Include your Python code and output in the code box below.)**
 Answer:





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# 1. Load a sample of the NYC Taxi fares dataset (assumes CSV with 'pickup_longitude', 'pickup_latitude')
df = pd.read_csv("nyc_taxi_fares_sample.csv", usecols=[
    "pickup_longitude", "pickup_latitude"], nrows=50000)
df.dropna(inplace=True)

# 2. Fit Isolation Forest
iso = IsolationForest(
    n_estimators=100,
    contamination=0.01,  # assume ~1% anomalies
    random_state=42
)
iso.fit(df)

# 3. Get anomaly predictions
df["anomaly"] = iso.predict(df)
# In scikit-learn: -1 = anomaly, 1 = inlier

# 4. Visualize results
fig, ax = plt.subplots(figsize=(8, 6))
inliers = df[df["anomaly"] == 1]
outliers = df[df["anomaly"] == -1]

ax.scatter(
    inliers["pickup_longitude"], inliers["pickup_latitude"],
    c="blue", s=2, label="Inliers"
)
ax.scatter(
    outliers["pickup_longitude"], outliers["pickup_latitude"],
    c="red", s=8, label="Anomalies"
)
ax.set_title("Isolation Forest Anomaly Detection on NYC Taxi Pickups")
ax.set_xlabel("Pickup Longitude")
ax.set_ylabel("Pickup Latitude")
ax.legend()
plt.show()
```

  **Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results. (Include your Python code and output in the code box below.)**
Answer:





```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 1. Load the dataset
url = "https://raw.githubusercontent.com/AileenNielsen/TimeSeriesAnalysisWithPython/master/data/AirPassengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
df.index.freq = 'MS'  # set monthly start frequency

# 2. Plot original series
df['#Passengers'].plot(figsize=(10, 4), title='Monthly Air Passengers (1949–1960)')
plt.ylabel('Passengers')
plt.show()

# 3. Fit SARIMA model
# Seasonal pattern = 12 months; after checking, using (p,d,q)x(P,D,Q,12)
model = SARIMAX(df['#Passengers'], order=(2,1,1), seasonal_order=(1,1,1,12),
                enforce_stationarity=False, enforce_invertibility=False)
result = model.fit(disp=False)
print(result.summary())

# 4. Forecast next 12 months
forecast = result.get_forecast(steps=12)
fc_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# 5. Plot forecast
plt.figure(figsize=(10, 5))
plt.plot(df['#Passengers'], label='Historical')
plt.plot(fc_mean.index, fc_mean, label='Forecast', color='darkgreen')
plt.fill_between(conf_int.index,
                 conf_int.iloc[:, 0],
                 conf_int.iloc[:, 1],
                 color='green', alpha=0.3)
plt.title('SARIMA Forecast – Next 12 Months of Air Passengers')
plt.xlabel('Date')
plt.ylabel('Passengers')
plt.legend()
plt.show()
```

 **Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib. (Include your Python code and output in the code box below.)**
  Answer:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# 1. Create synthetic 2D data with inliers and outliers
np.random.seed(42)
X_inliers = 0.3 * np.random.randn(200, 2)
X_inliers = np.r_[X_inliers + 2, X_inliers - 2]  # two dense clusters
X_outliers = np.random.uniform(low=-4, high=4, size=(40, 2))
X = np.r_[X_inliers, X_outliers]

# 2. Fit LOF for outlier detection
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)
# Labels: -1 = anomaly, 1 = inlier
scores = lof.negative_outlier_factor_

# 3. Visualize
plt.figure(figsize=(8, 6))
# Plot data points
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], c='blue', s=20, label='Inliers')
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], c='red', s=20, label='Anomalies')

# Overlay circles proportional to an “outlierness” score
radius = (scores.max() - scores) / (scores.max() - scores.min())
plt.scatter(X[:, 0], X[:, 1], s=1000 * radius, edgecolors='r', facecolors='none',
            label='Outlier score')

plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
